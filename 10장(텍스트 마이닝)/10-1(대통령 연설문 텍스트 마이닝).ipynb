```python
moon=open('speech_moon.txt',encoding='UTF-8').read()
moon
```

```python
import re
```

```python
moon=re.sub('[^가-힣]',' ',moon)   #가-힣 은 '한글이 아닌 모든 문자'를 의미. (불필요한 문자 제거)
```

```python
moon
```

```python
#명사 추출하기
```

```python
import konlpy
hannanum=konlpy.tag.Hannanum()
hannanum.nouns("대한민국의 영토는 한반도와 그 부속도서로 한다")   #이 문장에서 명사만 추출
```

```python
nouns=hannanum.nouns(moon)
nouns
```

```python
import pandas as pd
df_word=pd.DataFrame({'word':nouns})
df_word
```

```python
#4. 단어 빈도표 만들기
```

```python
df_word['count']=df_word['word'].str.len()   #단어의 글자 수 나타내는 변수 추가
df_word
```

```python
df_word=df_word.query('count>=2')    #두 글자 이상인 단어만 추출
df_word.sort_values('count')
```

```python
df_word=df_word.groupby('word',as_index=False).agg(n=('word','count')).sort_values('n',ascending=False)
```

```python
df_word
```

```python
top20=df_word.head(20)
```

```python
top20
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams.update({'font.family':'Malgun Gothic',
                    'figure.dpi':'120',
                    'figure.figsize':[6.5,6]})
sns.barplot(data=top20,x='n',y='word',hue='word')
```

```python
font='DoHyeon-Regular.ttf'         #한글 폰트 설정
```

```python
dic_word=df_word.set_index('word').to_dict()['n']        #데이터 프레임을 딕셔너리로 전환
dic_word
```

```python
from wordcloud import WordCloud
wc=WordCloud(random_state=1234,      #난수 설정
            font_path=font,         #폰트 설정
            width=400,               #가로 크기
            height=400,              #세로 크기
            background_color='black')    #배경색
```

```python
img_wordcloud=wc.generate_from_frequencies(dic_word)
```

```python
plt.figure(figsize=(10,10))   #가로, 세로 크기 설정
```

```python
plt.axis('off')     #테두리 선 없애기
```

```python
plt.imshow(img_wordcloud)        #워드 클라우드 출력
```

```python
import PIL
icon=PIL.Image.open('cloud.png')
```

```python
import numpy as np
img=PIL.Image.new('RGB',icon.size,(255,255,255))
img.paste(icon,icon)
img=np.array(img)
```

```python
wc=WordCloud(random_state=1234,
            font_path=font,
            width=400,
            height=400,
            background_color='white',
            mask=img)
```

```python
img_wordcloud=wc.generate_from_frequencies(dic_word)
```

```python
plt.figure(figsize=(10,10))
plt.axis('off')
plt.imshow(img_wordcloud)
```

```python
wc=WordCloud(random_state=1234,
            font_path=font,
            width=400,
            height=400,
            background_color='white',
            mask=img,
            colormap='inferno')         #컬러맵 설정
```

```python
img_wordcloud=wc.generate_from_frequencies(dic_word)
plt.figure(figsize=(10,10))
plt.axis('off')
plt.imshow(img_wordcloud)
```
