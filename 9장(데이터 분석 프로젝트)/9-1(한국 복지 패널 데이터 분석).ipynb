```python
##9-9(지역별 연령대 비율)
```

```python
welfare['code_region'].value_counts(
)
```

```python
list_region = pd.DataFrame({'code_region' : [1, 2, 3, 4, 5, 6, 7],
                            'region'      : ['서울',
                                             '수도권(인천/경기)',
                                             '부산/경남/울산',
                                             '대구/경북',
                                             '대전/충남',
                                             '강원/충북',
                                             '광주/전남/전북/제주도']})
list_region
```

```python
welfare=welfare.merge(list_region,how='left',on='code_region')
welfare[['code_region','region']].head()
```

```python
#1. 지역별 연령대 비율표 만들기
```

```python
region_ageg=welfare.groupby('region',as_index=False)['ageg'].value_counts(normalize=True)
region_ageg
```

```python
#2. 그래프 만들기
```

```python
region_ageg=region_ageg.assign(proportion=region_ageg['proportion']*100).round(1)
region_ageg
```

```python
sns.barplot(data=region_ageg,x='ageg',y='proportion',hue='region')
```

```python
sns.barplot(data=region_ageg,x='proportion',y='region',hue='ageg')
```

```python
#3. 누적 비율 막대 그래프 만들기
```

```python
#(1). 피벗하기: 행과 열을 회전해 표의 구성을 바꾸는 작업
```

```python
pivot_df=region_ageg[['region','ageg','proportion']].pivot(index='region',
                                                          columns='ageg',
                                                          values='proportion')
pivot_df
```

```python
#(2). 그래프 만들기
pivot_df.plot.barh(stacked=True)
```

```python
#(3). 막대 정렬하기
```

```python
#노년층 비율 기준 정렬, 변수 숫ㄴ서 바꾸기
reorder_df=pivot_df.sort_values('old')[['young','middle','old']]
reorder_df
```

```python
reorder_df.plot.barh(stacked=True)
```
