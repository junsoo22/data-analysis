```python
import pandas as pd
```

```python
mpg=pd.read_csv('mpg.csv')
mpg
```

```python
import seaborn as sns
sns.scatterplot(data=mpg,x='displ',y='hwy')   #x축은 displ, y축은 hwy를 나타낸 산점도
```

```python
sns.scatterplot(data=mpg,x='displ',y='hwy').set(xlim=[3,6])   #x축 범위 3~6으로 제한
```

```python
sns.scatterplot(data=mpg,x='displ',y='hwy').set(xlim=[3,6],ylim=[10,30])  #x축 범위 3~6, y축 범위 10~30으로 제한
```

```python
#drv별로 표식 색깔 다르게 표현(hue)
sns.scatterplot(data=mpg,x='displ',y='hwy',hue='drv')
```

```python
#그래프를 이미지 파일로 저장하기
#그래프에 마우스 커서 올리고 shift+마우스 우클릭
```

```python
#그래프 설정바꾸기
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.dpi' : '150'})             # 해상도, 기본값 72
plt.rcParams.update({'figure.figsize' : [8, 6]})        # 그림 크기, 기본값 [6, 4]
plt.rcParams.update({'font.size' : '15'})               # 글자 크기, 기본값 10
plt.rcParams.update({'font.family' : 'Malgun Gothic'})  # 폰트, 기본값 sans-serif
```

```python
#drv별로 표식 색깔 다르게 표현(hue)
sns.scatterplot(data=mpg,x='displ',y='hwy',hue='drv')
```
