```python
economics=pd.read_csv('economics.csv')
economics.head()
```

```python
import seaborn as sns
```

```python
sns.lineplot(data=economics, x='date',y='unemploy')  #date가 문자타입
```

```python
#날짜 시간 타입 변수 만들기
economics['date2']=pd.to_datetime(economics['date'])
economics.info()
```

```python
economics[['date','date2']]  #타입 바꾸더라도 값이 달라지지는 않음
```

```python
economics['date2'].dt.year   #연도 추출
```

```python
economics['date2'].dt.month   #월 추출
```

```python
economics['date2'].dt.day   #일 추출
```

```python
#연도 변수 추가
economics['year']=economics['date2'].dt.year
```

```python
economics[['year']]
```

```python
economics.head()
```

```python
sns.lineplot(data=economics,x='year',y='unemploy')   #x축에 연도 표시
```

```python
sns.lineplot(data=economics,x='year',y='unemploy',ci=None)   #신뢰구간 제거
```

```python
sns.lineplot(data=economics,x='year',y='unemploy',errorbar=None)   #신뢰구간 제거
```
