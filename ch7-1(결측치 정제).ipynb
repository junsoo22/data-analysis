```python
import pandas as pd
import numpy as np
```

```python
df=pd.DataFrame({'sex':['M','F',np.nan,'M','F'],   #결측치 만들기
             'score':[5,4,3,4,np.nan]})
df
```

```python
df['score']+1
```

```python
pd.isna(df)  #결측치 확인
```

```python
pd.isna(df).sum()   #결측치 빈도 확인
```

```python
df.dropna(subset=['score'])  #score 결측치 제거(결측치가 아닌 행만 출력)
```

```python
df_nomiss=df.dropna(subset=['score'])
df_nomiss['score']+1
```

```python
df_nomiss=df.dropna(subset=['score','sex'])
```

```python
df_nomiss
```

```python
df_nomiss2=df.dropna()   #모든 변수에 결측치 없는 데이터 추출(결측치 하나라도 있으면 제거)
df_nomiss2

```
